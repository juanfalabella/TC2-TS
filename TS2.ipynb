**1.** Hallar la transferencia $T= \frac{V_{o}}{V_{i}}$ en función de $ω_{o}$ y $Q$.

Transferencia hallada:

$ T(S) = \frac{V_{o}}{V_{i}} = -\frac{R_{3}}{R_{1}} . \frac{ \frac{1}{C^2 R_{3}^2}}{S^2 + S \frac{1}{C R_{2}} + \frac{1}{C^2 R_{3}^2}}  $

$ T(S) = \frac{V_{o}}{V_{i}} = K . \frac{ \omega_{o}^2}{S^2 + S \frac{\omega_{o}}{Q} + \omega_{o}^2} $

$ K = -\frac{R_{3}}{R_{1}} ;  \omega_{o} = \frac{1}{C R_{3}}    ;   Q = \frac{R_{2}}{R_{3}}  $

**3.** Ajustar el valor de $R_{1}$ de forma tal que $|T(0)|=20dB$.

$ |T(0)|_{dB} = 20 log( |T(0)|_{veces} ) $

$ 1 = log( |T(0)| ) $

$ |T(0)| = 10 = K $

$ K = \frac{R_{3}}{R_{1}} $

$ R_{1} = \frac{R_{3}}{K} = \frac{1}{10} $

+10 🎓 Calcular las sensibilidades  $S_{C}^{ω_{o}}$, $S_{R_{2}}^{Q}$ y $S_{R_{3}}^{Q}$

$S_{C}^{ω_{o}} = \frac{C  \partial ω_{o}}{ω_{o} \partial C} = C^{2}R_{3} \frac{-1}{C^{2}R_{3}} = -1$

$S_{R_{2}}^{Q} = \frac{R_{2}  \partial Q}{Q \partial R_{2}} = \frac{R_{3}*R_{2}}{R_{2}} \frac {1} {R_{3}}= 1$

$S_{R_{3}}^{Q} = \frac{R_{3}  \partial Q}{Q \partial R_{3}} = \frac{R_{3}*R_{3}}{R_{2}} \frac{-R_{2}}{R_{3}*R_{3}}= -1$
